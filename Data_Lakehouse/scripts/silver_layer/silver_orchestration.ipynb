###  This notebook runs all Silver transformation notebooks in sequence.
### It acts as the single entry point for the Silver layer in Databricks Jobs.



```python
notebooks = [
    "./CRM/silver_crm_cust_info",
    "./CRM/silver_crm_prd_info",
    "./CRM/silver_crm_sales_details",
    "./ERP/silver_erp_cust_az12",
    "./ERP/silver_erp_loc_a101",
    "./ERP/silver_erp_px_cat_g1v2"
]

success_list = []
failed_list = []

for nb in notebooks:
    try:
        print(f"Running {nb} ...")
        result = dbutils.notebook.run(nb, timeout_seconds=0)
        print(f"Successfully finished {nb}")
        success_list.append(nb)
    except Exception as e:
        print(f"Failed while running {nb}")
        print(f"Error: {str(e)}")
        failed_list.append(nb)
        raise

print("Succeeded notebooks:", success_list)
print("Failed notebooks:", failed_list)
```

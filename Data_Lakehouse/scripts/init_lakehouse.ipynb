#  Initial Lakehouse setup:
### Select the target catalog, create the Medallion layer schemas
### (Bronze, Silver, and Gold), and create a volume to store the raw source files.

# Select Catalog

```python
%sql
USE CATALOG workspace;
```

# Create Lakehouse Schemas

```python

%sql
CREATE SCHEMA IF NOT EXISTS bronze
COMMENT 'Bronze layer: raw ingested data';

CREATE SCHEMA IF NOT EXISTS silver
COMMENT 'Silver layer: cleaned and transformed data';

CREATE SCHEMA IF NOT EXISTS gold
COMMENT 'Gold layer: business-ready data';
```

# Create Volume

```python
%sql
CREATE VOLUME IF NOT EXISTS workspace.bronze.source_systems
COMMENT 'Volume for raw source files (CSV)';
```
